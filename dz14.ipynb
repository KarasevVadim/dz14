```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 12,9

# импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt
```

```python
konf = pd.read_csv('ucdp-prio-acd-4-2016.csv')
konf.head(10)
```

```python
a=konf.groupby('Year').count().reset_index()

a.head()
```

```python
plt.plot(a['Year'], a['ConflictId'])
plt.xlabel('Год')
plt.ylabel('Кол конфликтов')
plt.title('Кол военных конфликтов в указанном диапазоне лет')
```

```python
a.sort_values('ConflictId', ascending=False).head(1)
```
